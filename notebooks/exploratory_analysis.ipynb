# Netflix Dataset — Exploratory Data Analysis
This notebook explores the Netflix titles dataset used by the NetflixIQ dashboard.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

plt.style.use('dark_background')
sns.set_palette('husl')

df = pd.read_csv('../dataset/netflix_titles.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Basic info
print('Columns:', df.columns.tolist())
print('\nNull counts:')
print(df.isnull().sum())
print('\nType distribution:')
print(df['type'].value_counts())
```

```python
# Content type pie
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
type_counts = df['type'].value_counts()
axes[0].pie(type_counts, labels=type_counts.index, autopct='%1.1f%%',
            colors=['#E50914', '#00c9a7'], startangle=90)
axes[0].set_title('Movies vs TV Shows')

# Release year histogram
df_valid = df[(df['release_year'] > 1990) & (df['release_year'] <= 2022)]
axes[1].hist(df_valid['release_year'], bins=30, color='#E50914', alpha=0.8, edgecolor='black')
axes[1].set_title('Content by Release Year')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Top countries
top_countries = df[df['country'].notna() & (df['country'] != '')]['country'].value_counts().head(10)
plt.figure(figsize=(12, 5))
top_countries.plot(kind='barh', color='#E50914', edgecolor='black')
plt.title('Top 10 Content Producing Countries')
plt.xlabel('Number of Titles')
plt.tight_layout()
plt.show()
```

```python
# Genre analysis
all_genres = []
for g in df['listed_in'].dropna():
    all_genres.extend([x.strip() for x in g.split(',')])

genre_counts = Counter(all_genres).most_common(12)
genres, counts = zip(*genre_counts)

plt.figure(figsize=(12, 5))
plt.bar(genres, counts, color='#a855f7', edgecolor='black', alpha=0.85)
plt.xticks(rotation=35, ha='right')
plt.title('Top Content Genres')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print(f'\nTop genre: {genre_counts[0][0]} ({genre_counts[0][1]} titles)')
```

```python
# Ratings distribution
ratings = df[df['rating'].notna() & (df['rating'] != '')]['rating'].value_counts()
plt.figure(figsize=(10, 4))
ratings.plot(kind='bar', color='#f5c518', edgecolor='black', alpha=0.85)
plt.title('Content Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Movie duration analysis
movies = df[df['type'] == 'Movie'].copy()
movies['duration_mins'] = movies['duration'].str.extract(r'(\d+)').astype(float)
movies_valid = movies[movies['duration_mins'].notna()]

plt.figure(figsize=(10, 4))
plt.hist(movies_valid['duration_mins'], bins=25, color='#4fa3ff', edgecolor='black', alpha=0.85)
plt.axvline(movies_valid['duration_mins'].mean(), color='#E50914', linestyle='--', 
            linewidth=2, label=f'Mean: {movies_valid["duration_mins"].mean():.0f} min')
plt.title('Movie Duration Distribution')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()

print(f'Average movie duration: {movies_valid["duration_mins"].mean():.1f} minutes')
print(f'Median movie duration: {movies_valid["duration_mins"].median():.1f} minutes')
```

```python
# Summary statistics
print('=== NETFLIX DATASET SUMMARY ===')
print(f'Total Titles:  {len(df):,}')
print(f'Movies:        {len(df[df["type"]=="Movie"]):,}')
print(f'TV Shows:      {len(df[df["type"]=="TV Show"]):,}')
print(f'Year Range:    {int(df["release_year"].min())} – {int(df["release_year"].max())}')
print(f'Countries:     {df["country"].nunique()}')
print(f'Unique Genres: {len(set(all_genres))}')
```
